```python
from docx import Document

def generate_eviction_notice(output_path):
    doc = Document()

    # Header
    doc.add_paragraph(f"Date: [Insert Date]")
    doc.add_paragraph(f"[Landlord's Name]")
    doc.add_paragraph(f"[Landlord's Address]")
    doc.add_paragraph(f"[City, State, ZIP Code]\n")

    # Recipient Details
    doc.add_paragraph(f"[Tenant's Name]")
    doc.add_paragraph(f"[Tenant's Address]")
    doc.add_paragraph(f"[City, State, ZIP Code]\n")

    # Subject
    doc.add_paragraph("Subject: Eviction Notice for [Reason for Eviction]\n", style='Heading 1')

    # Introduction
    doc.add_paragraph(f"Dear [Tenant's Name],\n")
    doc.add_paragraph(
        "This letter serves as a formal notice to vacate the premises located at [Rental Property Address] due to [state the reason for eviction, e.g., non-payment of rent, violation of lease terms, expiration of lease, etc.]."
    )

    # Details of the Eviction Notice
    doc.add_paragraph("Details of the Eviction Notice:")
    doc.add_paragraph("1. Reason for Eviction: [Provide specific details, such as unpaid rent amount, lease violations, or other grounds for eviction.]")
    doc.add_paragraph("2. Lease Agreement: The lease agreement signed on [date] requires you to [state relevant lease terms].")
    doc.add_paragraph("3. Cure Period: [If applicable, state the number of days the tenant has to remedy the issue, e.g., pay overdue rent or correct a lease violation.]")
    doc.add_paragraph("4. Vacate Deadline: You are required to vacate the premises and return possession of the property by [date, typically 30 days from the notice date, depending on local laws].\n")

    # Consequences of Non-Compliance
    doc.add_paragraph("Consequences of Non-Compliance:")
    doc.add_paragraph(
        "If you fail to vacate the property by the specified deadline, legal action will be initiated to recover possession of the property. You may also be responsible for additional costs, including court fees, attorney fees, and any unpaid rent or damages.\n"
    )

    # Contact Information
    doc.add_paragraph("Contact Information:")
    doc.add_paragraph("If you have any questions or wish to discuss this matter further, please contact me at [your phone number] or [your email address].\n")

    # Closing
    doc.add_paragraph("Sincerely,")
    doc.add_paragraph("[Your Full Name]")
    doc.add_paragraph("[Your Signature]")

    # Save Document
    doc.save(output_path)
    print(f"Eviction notice template generated: {output_path}")

# Example Usage
generate_eviction_notice("eviction_notice_template.docx")
```

```python
from docx import Document

def generate_payment_notice(output_path):
    doc = Document()

    # Header
    doc.add_paragraph(f"Date: [Insert Date]")
    doc.add_paragraph(f"[Your Name]")
    doc.add_paragraph(f"[Your Address]")
    doc.add_paragraph(f"[City, State, ZIP Code]\n")

    # Recipient Details
    doc.add_paragraph(f"[Recipient's Name]")
    doc.add_paragraph(f"[Recipient's Address]")
    doc.add_paragraph(f"[City, State, ZIP Code]\n")

    # Subject
    doc.add_paragraph("Subject: Legal Notice for Payment of Outstanding Dues\n", style='Heading 1')

    # Introduction
    doc.add_paragraph(f"Dear [Recipient's Name],\n")
    doc.add_paragraph(
        "This legal notice is being sent to you in regard to the outstanding payment of [amount] due under [describe the agreement, invoice, or loan, e.g., Invoice No. 123, Loan Agreement dated 01/01/2023, etc.]."
    )

    # Details of Outstanding Dues
    doc.add_paragraph("Details of Outstanding Dues:")
    doc.add_paragraph("1. Amount Due: [amount]")
    doc.add_paragraph("2. Due Date: [due date]")
    doc.add_paragraph("3. Description: [brief description of the service, loan, or invoice, e.g., 'Payment for services rendered on 15/10/2023']")
    doc.add_paragraph("4. Reference: [reference number, invoice number, or agreement details]\n")

    # Request for Payment
    doc.add_paragraph("Request for Payment:")
    doc.add_paragraph(
        "Despite several reminders, the above-mentioned amount remains unpaid. We hereby request you to clear the outstanding dues of [amount] within [number of days, e.g., 7 or 15] days from the receipt of this notice."
    )

    # Consequences of Non-Payment
    doc.add_paragraph("Consequences of Non-Payment:")
    doc.add_paragraph(
        "Failure to make the payment within the stipulated time will compel us to initiate legal proceedings against you to recover the outstanding amount, along with any additional costs, interest, and legal fees incurred."
    )

    # Contact Information
    doc.add_paragraph("Contact Information:")
    doc.add_paragraph("If you have already made the payment, please disregard this notice. For any queries or clarification, please contact us at [your phone number] or [your email address].\n")

    # Closing
    doc.add_paragraph("Sincerely,")
    doc.add_paragraph("[Your Full Name]")
    doc.add_paragraph("[Your Signature]")

    # Save Document
    doc.save(output_path)
    print(f"Payment legal notice template generated: {output_path}")

# Example Usage
generate_payment_notice("payment_notice_template.docx")
```

```python
from docx import Document

def generate_property_dispute_notice(output_path):
    doc = Document()

    # Header
    doc.add_paragraph(f"Date: [Insert Date]")
    doc.add_paragraph(f"[Your Name]")
    doc.add_paragraph(f"[Your Address]")
    doc.add_paragraph(f"[City, State, ZIP Code]\n")

    # Recipient Details
    doc.add_paragraph(f"[Recipient's Name]")
    doc.add_paragraph(f"[Recipient's Address]")
    doc.add_paragraph(f"[City, State, ZIP Code]\n")

    # Subject
    doc.add_paragraph("Subject: Legal Notice for Property Dispute\n", style='Heading 1')

    # Introduction
    doc.add_paragraph(f"Dear [Recipient's Name],\n")
    doc.add_paragraph(
        "This legal notice is being issued to you in regard to the property located at [property address], which is legally owned by [your name] (hereinafter referred to as the 'Property'). It has come to our attention that you are [state the issue, e.g., encroaching on the property, illegally possessing the property, or disputing ownership]."
    )

    # Details of the Property
    doc.add_paragraph("Details of the Property:")
    doc.add_paragraph("1. Property Address: [property address]")
    doc.add_paragraph("2. Ownership Details: [brief description of ownership, e.g., 'Registered in my name under Document No. 123 dated 01/01/2023']")
    doc.add_paragraph("3. Nature of Dispute: [describe the dispute, e.g., boundary encroachment, illegal possession, inheritance claim, etc.]\n")

    # Assertion of Legal Rights
    doc.add_paragraph("Assertion of Legal Rights:")
    doc.add_paragraph(
        "I hereby assert my legal rights over the Property and demand that you immediately [state the demand, e.g., cease encroachment, vacate the property, or stop disputing ownership]. Any further unauthorized use, possession, or interference with the Property will be considered a violation of my legal rights."
    )

    # Consequences of Non-Compliance
    doc.add_paragraph("Consequences of Non-Compliance:")
    doc.add_paragraph(
        "Failure to comply with this notice within [number of days, e.g., 15 or 30] days from the receipt of this notice will compel me to initiate appropriate legal proceedings against you, including but not limited to filing a lawsuit for trespass, eviction, or recovery of possession. You may also be held liable for any damages, costs, and legal fees incurred."
    )

    # Contact Information
    doc.add_paragraph("Contact Information:")
    doc.add_paragraph("If you wish to resolve this matter amicably or have any queries, please contact me at [your phone number] or [your email address].\n")

    # Closing
    doc.add_paragraph("Sincerely,")
    doc.add_paragraph("[Your Full Name]")
    doc.add_paragraph("[Your Signature]")

    # Save Document
    doc.save(output_path)
    print(f"Property dispute notice template generated: {output_path}")

# Example Usage
generate_property_dispute_notice("property_dispute_notice_template.docx")
```

```python
from docx import Document

def generate_consumer_complaint_notice(output_path):
    doc = Document()

    # Header
    doc.add_paragraph(f"Date: [Insert Date]")
    doc.add_paragraph(f"[Your Name]")
    doc.add_paragraph(f"[Your Address]")
    doc.add_paragraph(f"[City, State, ZIP Code]\n")

    # Recipient Details
    doc.add_paragraph(f"[Company's Name]")
    doc.add_paragraph(f"[Company's Address]")
    doc.add_paragraph(f"[City, State, ZIP Code]\n")

    # Subject
    doc.add_paragraph("Subject: Legal Notice for Consumer Complaint\n", style='Heading 1')

    # Introduction
    doc.add_paragraph(f"Dear [Recipient's Name/Company's Name],\n")
    doc.add_paragraph(
        "This legal notice is being issued to you in regard to the [product/service] purchased by me on [date of purchase] from [place of purchase]. I regret to inform you that the [product/service] is defective/unsatisfactory and does not meet the standards promised by your company."
    )

    # Details of the Complaint
    doc.add_paragraph("Details of the Complaint:")
    doc.add_paragraph("1. Product/Service: [name of the product or service]")
    doc.add_paragraph("2. Date of Purchase: [date of purchase]")
    doc.add_paragraph("3. Invoice/Order Number: [invoice or order number]")
    doc.add_paragraph("4. Nature of Complaint: [describe the issue, e.g., 'The product is defective', 'The service was unsatisfactory', 'Misleading advertisement']")
    doc.add_paragraph("5. Defects/Issues: [list the specific defects or issues, e.g., 'The product stopped working after 2 days', 'The service was not as per the agreed terms']\n")

    # Redressal Request
    doc.add_paragraph("Redressal Request:")
    doc.add_paragraph(
        "In light of the above, I hereby request you to [state your demand, e.g., 'replace the product', 'refund the amount', 'provide a satisfactory resolution'] within [number of days, e.g., 7 or 15] days from the receipt of this notice. Failure to do so will leave me with no option but to escalate this matter further."
    )

    # Consequences of Non-Compliance
    doc.add_paragraph("Consequences of Non-Compliance:")
    doc.add_paragraph(
        "If you fail to address this issue within the stipulated time, I will be compelled to take appropriate legal action against your company, including but not limited to filing a complaint with the consumer court. You may also be held liable for any damages, costs, and legal fees incurred."
    )

    # Contact Information
    doc.add_paragraph("Contact Information:")
    doc.add_paragraph("If you wish to resolve this matter amicably or require further details, please contact me at [your phone number] or [your email address].\n")

    # Closing
    doc.add_paragraph("Sincerely,")
    doc.add_paragraph("[Your Full Name]")
    doc.add_paragraph("[Your Signature]")

    # Save Document
    doc.save(output_path)
    print(f"Consumer complaint notice template generated: {output_path}")

# Example Usage
generate_consumer_complaint_notice("consumer_complaint_notice_template.docx")
```
